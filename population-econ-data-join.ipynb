```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def transform_county(dfname, columnname):
    
    dfname[columnname] = dfname[columnname].str.split(' ').str[0]
    
    dfname[columnname] = dfname[columnname].str.lower()
    
```

```python
pop_2010 = pd.read_csv("population-econdata/2010_County_Population.csv", encoding = "ISO-8859-1")

transform_county(pop_2010, "County")

pop_2010.head()
```

```python
county_gdp = pd.read_csv("population-econdata/County_GDP.csv")

transform_county(county_gdp, "County")

county_gdp.head()
```

```python
zip_county = pd.read_csv("population-econdata/ZIP_County_FIPS.csv")

transform_county(zip_county, "COUNTYNAME")

zip_county.head()
```

```python
county_pov_income = pd.read_csv("population-econdata/County_Poverty_Income.csv")

transform_county(county_pov_income, "County_Name")

county_pov_income.head()
```

```python
df_merged = pd.merge(county_gdp, zip_county, on="STCOUNTYFP") 
df_merged = pd.merge(df_merged, pop_2010, on = "County")
df_merged = pd.merge(df_merged,county_pov_income, left_on = "County", right_on = "County_Name")

df_merged.head()
```

```python
df_merged.drop(columns=["State", "State_x", "State_y", "County",
                        "Unnamed: 3", "CLASSFP","County_Name"], 
              inplace = True)


df_merged.head()
```
